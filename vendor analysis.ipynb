```python
df['Volume']=df['Volume'].astype('float64')
```

```python
df.fillna(0,inplace=True)
```

```python
df['VendorName']=df['VendorName'].str.strip()
```

```python
df['gross_profit']=df['total_sales_dollars']-df['total_purchase_dollars']
```

```python
df
```

```python
df['gross_profit'].min()
```

```python
 df['profit_margin']=df['gross_profit']/df['total_sales_dollars']*100
```

```python
df['stock_turnover']=df['total_sales_quantity']/df['total_purchase_quantity']
```

```python
df['sales_to_purchase_ratio']=df['total_sales_dollars']/df['total_purchase_dollars']
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine(
    "postgresql+psycopg2://postgres:qwerty@localhost:5432/vendor_performance"
)

df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

df.to_sql(
    "final_table",
    con=engine,
    if_exists="replace",
    index=False
)
```
